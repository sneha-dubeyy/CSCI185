```python
#Q1 Code:

data = pd.read_csv("boston.csv")
print("Columns: ", data.columns, "\n")

dataToAnalyse = data.drop('Unnamed: 0', axis=1)

correlations = dataToAnalyse.corr()

for i in data.columns[1:]:
    for j in data.columns[1:]:
        if i != j and abs(correlations[i][j]) > 0.8:
            print(i, " and ", j, " are highly correlated with CC = ", correlations[i][j], "\n")

print('\n')
print('Correlation Matrix:')       
correlations
```

```python
#Q2, Part A

dataFrame = pd.DataFrame(fetch_california_housing().data, columns = fetch_california_housing().feature_names)
print('Data: ', dataFrame)
```

```python
#Q2, Part B

#Min-Max
custom_range = (0, 10)

scaler = MinMaxScaler(feature_range=custom_range)

mmNormalisedData = scaler.fit_transform(dataFrame)

print('normalised using min-max: ', mmNormalisedData)

#Z-Score Normalziation

scaler = StandardScaler()
zNormalisedData = scaler.fit_transform(dataFrame)

print('normalised using z-score: ', zNormalisedData)
```
